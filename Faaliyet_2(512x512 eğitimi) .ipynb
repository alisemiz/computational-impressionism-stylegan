```python
%%writefile resize_script.py
# Bu hÃ¼crenin tÃ¼m iÃ§eriÄŸi 'resize_script.py' adÄ±yla kaydedilecek
# Bu script, 'pillow' ve 'tqdm' kÃ¼tÃ¼phanelerinin kurulu olduÄŸu 'myenv' ortamÄ± tarafÄ±ndan Ã§alÄ±ÅŸtÄ±rÄ±lacak.

import os
import zipfile
from PIL import Image
from tqdm import tqdm
import sys

# YollarÄ± tanÄ±mlayalÄ±m
SOURCE_ZIP = "/content/drive/MyDrive/TUBITAK_Projesi/dataset_temiz_v1.zip"
TEMP_DIR = "/content/drive/MyDrive/TUBITAK_Projesi/empresyonizm_512_temp_klasoru"
TARGET_SIZE = (512, 512)

print(f"--- 'resize_script.py' BaÅŸlatÄ±lÄ±yor ---")
print(f"Kaynak: {SOURCE_ZIP}")
print(f"Hedef KlasÃ¶r: {TEMP_DIR}")

# GeÃ§ici klasÃ¶rÃ¼ oluÅŸturalÄ±m
os.makedirs(TEMP_DIR, exist_ok=True)

# 'dataset_temiz_v1.zip' dosyasÄ±nÄ± aÃ§ ve iÅŸle diyelim
try:
    with zipfile.ZipFile(SOURCE_ZIP, 'r') as zf:
        # GeÃ§erli resim dosyalarÄ±nÄ±n listesini alalÄ±m
        file_list = [f for f in zf.namelist() if not f.startswith('__') and f.lower().endswith(('.png', '.jpg', '.jpeg'))]

        # tqdm ile bir ilerleme Ã§ubuÄŸu oluÅŸturalÄ±m
        for filename in tqdm(file_list, desc="Resimler 512x512'ye dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lÃ¼yor"):
            try:
                with zf.open(filename) as f:
                    # Resmi Pillow ile aÃ§ ve RGB'ye dÃ¶nÃ¼ÅŸtÃ¼r diyelim (eÄŸitim iÃ§in Ã¶nemli)
                    img = Image.open(f).convert('RGB')
                    # Yeniden boyutlandÄ±r diyelim (LANCZOS en yÃ¼ksek kaliteyi verir)
                    img_resized = img.resize(TARGET_SIZE, Image.LANCZOS)
                    # Yeni resmi geÃ§ici klasÃ¶re kaydedelim
                    save_path = os.path.join(TEMP_DIR, os.path.basename(filename))
                    img_resized.save(save_path)
            except Exception as e:
                print(f"Hata (dosya atlanÄ±yor: {filename}): {e}")

    print(f"\nBaÅŸarÄ±lÄ±! {len(file_list)} resim {TEMP_DIR} klasÃ¶rÃ¼ne kaydedildi.")

except FileNotFoundError:
    print(f"KRÄ°TÄ°K HATA: {SOURCE_ZIP} dosyasÄ± bulunamadÄ±. LÃ¼tfen dosya yolunu kontrol edin.", file=sys.stderr)
    sys.exit(1)
except Exception as e:
    print(f"KRÄ°TÄ°K HATA (Zip dosyasÄ± okunamadÄ±): {e}", file=sys.stderr)
    sys.exit(1)

print("\n--- 'resize_script.py' TamamlandÄ± ---")
```
