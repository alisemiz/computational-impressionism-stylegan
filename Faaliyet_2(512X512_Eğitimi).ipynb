```python
import os
import glob
import shutil
import sys
import zipfile
from PIL import Image
from tqdm import tqdm
import io

print("--- HÜCRE 2: Dosyalar Hazırlanıyor ---")

# 1. MODELİ OTOMATİK BULMA
print("\n1. En son eğitilmiş model aranıyor...")
# Tüm results_512 klasörlerini tara
base_path = "/content/drive/MyDrive/TUBITAK_Projesi/results_512*"
# Tüm .pkl dosyalarını bul
all_pkls = glob.glob(os.path.join(base_path, "**", "network-snapshot-*.pkl"), recursive=True)

if len(all_pkls) > 0:
    # İsme göre sırala (En yüksek sayı en sonda olur)
    all_pkls.sort()
    latest_pkl = all_pkls[-1] # En son dosya

    print(f"En güncel model bulundu: {os.path.basename(latest_pkl)}")
    print(f"Yol: {latest_pkl}")

    # Kopyala
    local_resume_pkl = "/content/resume.pkl"
    shutil.copy(latest_pkl, local_resume_pkl)
    print("Model yerel diske kopyalandı (/content/resume.pkl)")
else:
    print("HATA: Hiçbir .pkl dosyası bulunamadı! Lütfen Drive'ı kontrol edin.")
    sys.exit()

# 2. VERİ SETİNİ HAZIRLAMA
-
print("\n2. Veri seti hazırlanıyor (512x512)...")
raw_data_zip = "/content/drive/MyDrive/TUBITAK_Projesi/dataset_temiz_v1.zip"
ready_data_zip = "/content/empresyonizm_512_ready.zip"
TARGET_SIZE = (512, 512)

if not os.path.exists(raw_data_zip):
    print("HATA: Veri seti bulunamadı!")
    sys.exit()

try:
    with zipfile.ZipFile(raw_data_zip, 'r') as zf_in:
        with zipfile.ZipFile(ready_data_zip, 'w', compression=zipfile.ZIP_STORED) as zf_out:
            file_list = [f for f in zf_in.namelist() if not f.startswith('__') and f.lower().endswith(('.png', '.jpg', '.jpeg'))]
            for filename in tqdm(file_list, desc="Dönüştürülüyor"):
                with zf_in.open(filename) as f:
                    img = Image.open(f).convert('RGB')
                    img_resized = img.resize(TARGET_SIZE, Image.LANCZOS)
                    img_byte_arr = io.BytesIO()
                    img_resized.save(img_byte_arr, format='PNG')
                    zf_out.writestr(os.path.basename(filename) + ".png", img_byte_arr.getvalue())
    print("\nVeri seti hazırlandı.")
except Exception as e:
    print(f"Veri hatası: {e}")
    sys.exit()
```
